```python
# B for Brown, G for Green, C for Clear
bottle_type = ['B', 'G', 'C']

# The are 6 permutations of bottle mapping
permutation_function = [[0, 1, 2],
                        [0, 2, 1],
                        [1, 0, 2],
                        [1, 2, 0],
                        [2, 0, 1],
                        [2, 1, 0]]

# The total movement
min_movement_count = 0

# The result of movement
min_movement_bin = ""
```

```python
def ecological_bin_packing(init_bin):
    global min_movement_count, min_movement_bin
    
    # bottles[i][j] means the number of type j bottles in bin i
    bottle_in_bin = []
    
    for i in range (3):
        current_bin = []
        for j in range (3):
            current_bin.append(init_bin[j+(i*3)])
        bottle_in_bin.append(current_bin)
    print(bottle_in_bin)
    
    for i in range (len(permutation_function)):
        current_movement_count = 0
        for j in range (3):
            for k in range (3):
                if permutation_function[i][j] != k:
                    current_movement_count += bottle_in_bin[j][k]
                    print(permutation_function[i][j], bottle_in_bin[j][k], current_movement_count)
            print()
        
        current_movement_bin = bottle_type[permutation_function[i][0]] + bottle_type[permutation_function[i][1]] + bottle_type[permutation_function[i][2]];
        print("bin = ", current_movement_bin)
#       or (min_movement_count == current_movement_count and current_movement_bin.compareTo(min_movement_bin) < 0)
        if (min_movement_count == 0) or (min_movement_count > current_movement_count):
            min_movement_count = current_movement_count
            min_movement_bin = current_movement_bin
            print("min = ", min_movement_count)
            print()
            print()
            print()
    
    print(min_movement_count,"-", min_movement_bin)
```

```python
init_bin = [1, 2, 3, 4, 5, 6, 7, 8, 9]
ecological_bin_packing(init_bin)
```
